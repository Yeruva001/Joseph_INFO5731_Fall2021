```python
# Write your code here

# Scraping the Review Content from Amazon

''' Importing Beautiful Soup Library for Web Scraping and requests to get data from the URL'''

from bs4 import BeautifulSoup
import requests

cust_request = requests.get("https://www.amazon.in/Apple-iPhone-11-128GB-Black/product-reviews/B07XVLW7YK/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews")

''' To print the URL that we search for along with Query '''
print(cust_request.url)

''' This will print the HTML content of the web page which doesn't look like HTML format though but I further use the prettify() function to print the content in HTML format. '''

print(cust_request.content)

```

```python
''' We need to provide HTML parser by default. '''
soup_Beau = BeautifulSoup(cust_request.text, 'html.parser')

''' To improve the appearance of the HTML code, use prettify.'''
print(soup_Beau.prettify())

prof_Name = soup_Beau.findAll("span", {"class" : "a-profile-name"})     # find <span> element with attributes class = a-profile-name

# Name extracted from span tag
prof_Name_Rev = []
for i in range(2,len(prof_Name)):     # We must exclude the Top Positive and Top Critical Review, which would otherwise be repeated, so the loop is started at step 2.
    prof_Name_Rev.append(prof_Name[i].get_text())
    
print(prof_Name_Rev)
```

```python
prod_Rat = soup_Beau.findAll("i", {"class" : "review-rating"})     # find <i> element with attributes class = review-rating

# Extract name from span tag
prod_Rat_Rev = []
for i in range(2,len(prod_Rat)):     # We must exclude the Top Positive and Top Critical Review, which would otherwise be repeated, so the loop is started at step 2.
    prod_Rat_Rev.append(prod_Rat[i].get_text())
    
print(prod_Rat_Rev)
```

```python
tit_Rev = soup_Beau.findAll("a", {"class" : "review-title-content"})     # find <a> element with attributes class = review-title-content
```

```python
# Name extracted from span tag
sum_Rev = []
for i in range(0,len(tit_Rev)):
    sum_Rev.append(tit_Rev[i].get_text())

# We shall remove the '\n' from before and after of every Review Title
sum_Rev[:] = [i.lstrip('\n').rstrip('\n') for i in sum_Rev]

print(sum_Rev)
```

```python
des_Rev = soup_Beau.findAll("span", {"class" : "review-text-content"})     # find <span> element with attributes class = review-text-content

# Extract name from span tag
des_Nam = []
for i in range(0,len(des_Rev)):
    des_Nam.append(des_Rev[i].get_text())

# We will remove the '\n' from before and after of every Review Desciption
des_Nam[:] = [i.lstrip('\n').rstrip('\n') for i in des_Nam]

print(des_Nam)
```

```python
# Saving the Scraped Content to .csv file from

import pandas as pd

# Make a DataFrame to add Details in it
cust_Rev_Data_Fram = pd.DataFrame()

# Adding Review Information in DataFrame
cust_Rev_Data_Fram["Reviewer"] = prof_Name_Rev
cust_Rev_Data_Fram["Rating"] = prod_Rat_Rev
cust_Rev_Data_Fram["Review Title"] = sum_Rev
cust_Rev_Data_Fram["Review Description"] = des_Nam

cust_Rev_Data_Fram
```

```python
# Converting DataFrame to .csv file and Saving it in your Machine

cust_Rev_Data_Fram.to_csv(r'customerReviews.csv', index= 'True')
```

```python
import spacy
nlp = spacy.load('en_core_web_sm') 

noun = []
adjective = []
verb = []
adverb = []

for w in tot_sent: 
  wd = nlp(w)
  for t in wd:
    if t.pos_ == 'NOUN': 
      noun.append(t)
    if t.pos_ == 'ADJ': 
      adjective.append(t)
    if t.pos_ == "VERB": 
      verb.append(t)
    if t.pos_ == "ADV": 
      adverb.append(t)
print("Adverb Count: ",len(adverb))
print("Adjective Count: ",len(adjective))
print("Verb Count: ",len(verb))
print("Noun Count: ",len(noun))
```
