```python
# Write your code here
import pandas as pd
i=pd.read_csv('https://raw.githubusercontent.com/Yeruva001/Joseph_INFO5731_Fall2021/main/data1.csv')
print(i.shape)


```

```python
# Write your code here


import pandas as pd
i=pd.read_csv('https://raw.githubusercontent.com/Yeruva001/Joseph_INFO5731_Fall2021/main/data1.csv')
print(i.shape)


```

```python
import nltk
i['Cleaned Data'] = i['Cleaned Data'].str.replace('[^\w\s]','')  
i['Cleaned Data'] = i['Cleaned Data'].apply(lambda x: " ".join(x.lower() for x in x.split()))
from nltk.corpus import stopwords 
nltk.download('stopwords')
s = stopwords.words('english')
i['Cleaned Data'] = i['Cleaned Data'].apply(lambda x: " ".join(x for x in x.split() if x not in s))
```

```python
i.head()
```

```python
i.info()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tf_i= TfidfVectorizer(ngram_range=(1,2), max_features=1000)
tf_i.fit(i['Cleaned Data'])
x_v =  tf_i.transform(i['Cleaned Data'])
y_v = i['Sentimental Analysis']


# Split the training data to training and validating data
from sklearn.model_selection import train_test_split
x_tra, x_val, y_tra, y_val = train_test_split(x_v, y_v, test_size = 0.2)
```

```python
from sklearn import metrics

def eval(y_prd, y_tst):
    Acc = metrics.accuracy_score(y_prd, y_tst)
    Re = metrics.recall_score(y_pred = y_prd, y_true = y_tst, pos_label='positive', average='micro') 
    Pre = metrics.precision_score(y_pred = y_prd, y_true = y_tst, pos_label='positive', average='micro') 
    F = 2 * (Pre * Re) / (Pre + Re)
    print("Accuracy: ", Acc.round(4))
    print("Recall:", Re.round(4))
    print("Precision:", Pre.round(4))
    print("F-1 score:", F.round(4))
```

```python
from sklearn import naive_bayes
from sklearn.model_selection import cross_val_score, KFold

na_imp = naive_bayes.MultinomialNB()

na_imp.fit(x_tra,y_tra)

y_pre_val = na_imp.predict(x_val)
eval(y_pre_val, y_val)
print("Cross Validation Score:", cross_val_score(na_imp, x_val, y_val, cv= KFold(10, shuffle=True, random_state = 22)))
```

```python
from sklearn import svm

svm_imp = svm.SVC()

svm_imp.fit(x_tra,y_tra)

y_pre_val = svm_imp.predict(x_val)
eval(y_pre_val, y_val)
print("Cross Validation Score:", cross_val_score(svm_imp, x_val, y_val, cv= KFold(10, shuffle=True, random_state = 22)).mean().round(4))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import numpy as np

df_tra = df_tra.select_dtypes(include=['number']).interpolate().dropna()
df_tst = df_tst.select_dtypes(include=['number']).interpolate().dropna()

x_tra_data = df_tra.drop(['SalePrice','Id'], axis=1)
y_tra_data = np.log(df_tra.SalePrice)


x_tra, x_tst, y_tra, y_tst = train_test_split(x_tra_data,y_tra_data,random_state = 21, test_size=0.2)


reg = LinearRegression()
reg.fit(x_tra,y_tra)

y_prd = reg.predict(x_tst)
```

```python
print('Linear Regression": %.4f' % reg.score(x_tst, y_tst))
```

```python
import numpy as np
from sklearn.metrics import mean_squared_error
l = mean_squared_error(np.exp(y_prd), y_tst)
li_r = np.sqrt(l)
print(li_r)
```
