Question 2 (20 points): Write python code to extract these features you discussed above. You can collect a few sample text data for the feature extraction. 

```python
# You code here (Please add comments in the code):


import nltk

from bs4 import BeautifulSoup as bs
import pandas as pd
import requests as rq
import matplotlib.pyplot as plt
import nltk
from wordcloud import WordCloud
from nltk.corpus import stopwords

#collecting sample text data from IMDb "KGF2 movie reviews"

url1 = 'https://www.imdb.com/title/tt10698680/reviews?ref_=tt_ov_rt'

#Storing the reviews into a dataframe
df1 = pd.DataFrame(columns=['Review'])
reqest = rq.get(url1).text
soup1 = bs(reqest,'html.parser')

#collecting reviews as text
rev_txt = soup1.find_all('div', attrs={'class' : 'text show-more__control'})

#taking a empty list initially
rev_lst = []
for i in range(len(rev_txt)):
    rev_lst.append(rev_txt[i].get_text())

#assigining list of reviews to the data frame    
df1['Review'] = rev_lst

df1['word_count'] = df1['Review'].apply(lambda x: len(str(x).split(" ")))
df1[['Review','word_count']].head()
```
